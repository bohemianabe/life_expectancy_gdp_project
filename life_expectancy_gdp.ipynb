```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
df.rename(columns={'Life expectancy at birth (years)': 'life_expectancy'}, inplace=True)
print((len(df)))
print(df.head())
```

```python
countries = [i for i in df.Country.unique()]
print(countries)
```

```python
chile = df[df.Country == 'Chile']
china = df[df.Country == 'China']
germany = df[df.Country == 'Germany']
mexico = df[df.Country == 'Mexico']
usa = df[df.Country == 'United States of America']
zimbabwe = df[df.Country == 'Zimbabwe']
print(chile)
```

```python
plt.figure(figsize=(10, 8))
plt.title('Life Expectancy 2010 - 2017')
sns.boxplot(data=df, x='Country', y='life_expectancy')
plt.savefig('box_lifeExpectancyALL')
plt.show()
```

```python
# on the list of countries analyzed here zimbabwe is an extreme outlier and need to be examined more closely if not also
# separately
```

```python
df_2 = pd.concat([chile, china, germany, mexico, usa])
plt.title('Life Expectancy 2010-2017')
sns.boxplot(data=df_2, x='Country', y='life_expectancy')
plt.savefig('box_lifeExpectancy')
plt.show()
```

```python
# chile and germany outperform all other countries
```

```python
plt.figure(figsize=(10, 8))
plt.title("Country's GDP 2010-2017")
sns.barplot(data=df, x='Country', y='GDP')
plt.savefig('country_and_GDP.png')
plt.show()
```

```python
# noticable is the larger GDP of the USA, however it compared to the box plot above does not seem to outperform
# the other nations (minus zimbabwe) as significantly. Is there a saturation point where GDP growth is negligable 
# on the outcome of life expectancy? 
```

```python
plt.figure(figsize=(15, 8))
plt.subplots_adjust(wspace=0.35, hspace=.40)
ax1 = plt.subplot(2, 3, 1)
ax1 = plt.plot(zimbabwe.life_expectancy, zimbabwe.GDP, marker='o')
ax1 = plt.title('Zimbabwe Life Expectancy')
ax1 = plt.xlabel('Life Expectancy')
ax1 = plt.ylabel('GDP')

ax2 = plt.subplot(2, 3, 2)
ax2 = plt.plot(china.life_expectancy, china.GDP, marker='o')
ax2 = plt.title('China Life Expectancy')
ax2 = plt.xlabel('Life Expectancy')
ax2 = plt.ylabel('GDP')

ax3 = plt.subplot(2, 3, 3)
ax3 = plt.plot(germany.life_expectancy, germany.GDP, marker='o')
ax3 = plt.title('German Life Expectancy')
ax3 = plt.xlabel('Life Expectancy')
ax3 = plt.ylabel('GDP')

ax4 = plt.subplot(2, 3, 4)
ax4 = plt.plot(mexico.life_expectancy, mexico.GDP, marker='o')
ax4 = plt.title('Mexican Life Expectancy')
ax4 = plt.xlabel('Life Expectancy')
ax4 = plt.ylabel('GDP')

ax5 = plt.subplot(2, 3, 5)
ax5 = plt.plot(usa.life_expectancy, usa.GDP, marker='o')
ax5 = plt.title('US Life Expectancy')
ax5 = plt.xlabel('Life Expectancy')
ax5 = plt.ylabel('GDP')

ax6 = plt.subplot(2, 3, 6)
ax6 = plt.plot(chile.life_expectancy, chile.GDP, marker='o')
ax6 = plt.title('Chile Life Expectancy')
ax6 = plt.xlabel('Life Expectancy')
ax6 = plt.ylabel('GDP')
plt.savefig('line_lifeExpectancy_vs_GDP')
plt.show()
```

```python
# here we see that life expectancy doesn't always correlate to the growth of GDP with some fluctuations. 
# I wonder if there were certain local events that outside of our df that could affect life expectancy such as a recession
# or other event?
```
